ye sara kam bhi function ke through hi hota hai

```python
def max_runs_against_team(bats_man):
  VK = delivery[delivery['batsman']== bats_man ]
  team = VK.groupby('bowling_team')
  return  team['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
max_runs_against_team('V Kohli')
```

```python
max_runs_against_team('A Ashish Reddy')
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
x = all_batsman>200
batsmanX = all_batsman[x].index.tolist()
batsmanX
```

```python
AvailableBats = death_over['batsman'].isin(batsmanX) #final = delivery[delivery['batsman'].isin(batsman_list)       sir ne delivery pr laga diya hai vedio mein ek comment diya hai ki death_over pr lagega masking
final = death_over[AvailableBats]
runs = final.groupby('batsman')['batsman_runs'].sum()
bolls = final.groupby('batsman')['batsman_runs'].count()
strickRate = (runs/bolls)*100
strickRate.sort_values(ascending=False)
```
